```python
import numpy as np
import helper
from helper import InterpolateFonts
```

Usage:

  * `font = InterpolateFonts(font1, font2)`
  * `font.interpolate(alpha=0.5)` - will give the middle between the two fonts, can change the alpha to be any number. 0 is font1, 1 is font 2
  * `font.add(w=0)` - will add w*font2 to font1. w can be any number
  
Additional methods:

  * `show_char(font, char)` - shows the real and predicted characters, side by side. E.g. `show_char(160, 3)` will show the the letter D (index 3) of font number 160. 
  * `show_real_font(font)` - will show all 62 chars of a real font (chosen by its index number)
  * `show_predicted_font(font)` - will show all 62 chars of the predicted font (chosen by its index number)
